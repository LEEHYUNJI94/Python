*     BeautifulSoup-원하는 부분 긁어내기1


```
import requests,json
from bs4 import BeautifulSoup

response = requests.get("https://askdjango.github.io/lv1/")
print(response)
print(response.status_code)
print(response.text)
```

```
#응답 html코드를 text로 변환#
html=response.text

#응답받은 html코드를 BeautifulSoup에 사용하기 위해 인스턴스지정#
soup1=BeautifulSoup(html,'html.parser')          #html관점에서 문자열 이해하기

#원하는 태그 지정해서 출력# 
for tag in soup1.select("li[class=course]"):     #html태그에 class=course만 select
  print(tag.text)
  print(soup1.head.title)                        #html태그에 title태그만 출력
for link in soup1.find_all('a'):                 #a태그로 둘러싸인 텍스트와 a태그의 href속성 출력
   print(link.text.strip(), link.get('href'))    #strip() : 문자열 좌우의 공백 문자를 없앤 것을 반환한다
```

*     BeautifulSoup-원하는 부분 긁어내기2

```
import urllib.request , urllib.parse, urllib.error
from bs4 import BeautifulSoup

#웹 주소 받기#
url=input("실행할 주소를 입력해주세요:")
html=urllib.request.urlopen(url).read()
soup=BeautifulSoup(html,'html.parser')

#웹 링크 크롤링하기#
tags = soup('a')
for tag in tags :
  print(tag.get('href',None))
```
